```python
import requests
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
from datetime import datetime
from meteostat import Point, Daily

```

```python
# Location: New York City (latitude, longitude, altitude)
location = Point(40.7128, -74.0060, 10)
start = datetime(2023, 1, 1)
end = datetime(2024, 12, 31)
# Get daily data for the location between start and end dates
data = Daily(location, start, end)

# Fetch the data into a pandas DataFrame
df = data.fetch()
df.head()
```

```python
df.info()
```

```python
df.reset_index(inplace = True)
```

```python
colums_to_keep = ['time','tmin', 'tmax', 'prcp', 'wspd', 'pres','tavg']
data = df[colums_to_keep]

```

```python
# Reanamming the columns
# Rename columns with full descriptive names
data.rename(columns={
    'tmin': 'Minimum_Temperature_C',
    'tmax': 'Maximum_Temperature_C',
    'prcp': 'Precipitation_mm',
    'wspd': 'Wind_Speed_km_h',
    'pres': 'Atmospheric_Pressure_hPa',
    'tavg': 'Average_Temperature_C'
} ,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
# Extract features
data['Month'] = data['time'].dt.month
data['Day'] = data['time'].dt.day
data['Day_of_Week'] = data['time'].dt.dayofweek

# Optional: drop 'time' if you don’t need it anymore
# data.drop('time', axis=1, inplace=True)

# Check
data.head()
```

```python

# Features
X = data[['Minimum_Temperature_C', 'Maximum_Temperature_C',
          'Precipitation_mm', 'Wind_Speed_km_h', 'Atmospheric_Pressure_hPa',
          'Month', 'Day', 'Day_of_Week']]

# Target
Y = data['Average_Temperature_C']

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    Y, 
    test_size=0.2,       # 20% test, 80% train (common choice)
    random_state=42      # ensures reproducibility
)
```

```python
# Training the model linear regression 
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
# Evaluate performance
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("Linear regression ")
print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
# Trainnig lasso model for this too
lasso = Lasso(alpha = 0.1)
lasso.fit(X_train,y_train)
y_pred = lasso.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Lasso Regression Results:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
# Now i want to predict on the behalf of RandomForestRegressor
# Initialize the model
rf = RandomForestRegressor(
    n_estimators=100,      # number of trees (default = 100)
    random_state=42,       # ensures reproducibility
    max_depth=None,        # you can limit tree depth if overfitting
    n_jobs=-1              # use all CPU cores for faster training
)

# Train the model
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Evaluate performance
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Random Forest Regression Results:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
# Set Seaborn style for cleaner visuals
sns.set(style="whitegrid", context="talk")

plt.figure(figsize=(10,7))

# Scatter plot of actual vs predicted
sns.scatterplot(x=y_test, y=y_pred, alpha=0.7, color="#1f77b4", edgecolor="k")

# Perfect prediction line
lims = [min(y_test.min(), y_pred.min()), max(y_test.max(), y_pred.max())]
plt.plot(lims, lims, 'r--', linewidth=2, label='Perfect Prediction')

# Add labels and title
plt.xlabel("Actual Temperature (°C)", fontsize=14)
plt.ylabel("Predicted Temperature (°C)", fontsize=14)
plt.title("🌦️ Actual vs Predicted Temperature (Lasso Regression)", fontsize=16, weight='bold')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(y_test.values[:100], label='Actual', marker='o', linewidth=2)
plt.plot(y_pred[:100], label='Predicted', marker='x', linewidth=2)
plt.title("Temperature Prediction Trend (First 100 Samples)", fontsize=16, weight='bold')
plt.xlabel("Sample Index")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
